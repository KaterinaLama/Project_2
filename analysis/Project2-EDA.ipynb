```python
# Find index of max value
max_idx = df['Total_emissions'].idxmax()
min_idx = df['Total_emissions'].idxmin()

# Build color list
# Build color list
colors = []
for i in df.index:
    if i == max_idx:
        colors.append('red')   # highest
    elif i == min_idx:
        colors.append('green')    # lowest
    else:
        colors.append('lightblue')     # others

plt.figure(figsize=(10, 6))  # wider and taller
plt.bar(df['Food product'], df['Total_emissions'], color=colors)
plt.xlabel("Food Product")
plt.ylabel("Total Emissions CO2")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
